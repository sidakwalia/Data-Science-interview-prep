```python
#QGeneratin random numbers

#(a:0.5,b:0.2,0.3)
import numpy as np
```

```python
lis1=[]
for ix in range(1000):                                             
    x=np.random.random()  #This function generates random value between [0,1)
    if(x<=0.5):              #generates values <=0.5
        lis1.append("a")
    if(x>0.5 and x<=0.7):    #generates values>0.5 && <=0.7
        lis1.append("b")
    if(x>0.7):                #generates values >0.7
        lis1.append("c")
```

```python
x=np.random.multivariate_normal([1.0,0.5],[[1,-1.5],[-1.5,1.5]],1000)  #normal distribution of the data
```

```python
plt.scatter(x[:,0],x[:,1])               
```

```python
stdnormaldis=(x-x.mean())/x.std()      #standard normal distribution
```

```python
plt.scatter(stdnormaldis[:,0],stdnormaldis[:,1],c='b')
```

```python
#create and plot frequency of each title length
```

```python
m=movies["movie_title"].values
title_length=[len(ix) for ix in m]    #calculates the length of the movies
```

```python
len_dict={} #create a dictoniary storing all the length

for ix in title_length:
    if ix in len_dict.keys():
        len_dict[ix]+=1
    else:
        len_dict[ix]=1


```

```python
plt.bar(range(len(len_dict)),list(len_dict.values()))  #Because earlier we were comparing a string with the lenght
plt.show()
```
