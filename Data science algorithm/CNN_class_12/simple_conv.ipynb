```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("doctor-strange-2-benedict-cumberbatch.jpg", 0)

print (img.shape)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
kernel = []
kernel.append([-1.0,-1.0,-1.0])
kernel.append([-1.0,8.0,-1.0])
kernel.append([-1.0,-1.0,-1.0])

kernel = np.array(kernel)

print (kernel)
```

```python
new_img = np.zeros((img.shape[0]-kernel.shape[0]+1, img.shape[1]-kernel.shape[1]+1))
print (new_img.shape)
```

```python
## convolution

for ix in range(new_img.shape[0]):
    for ij in range(new_img.shape[1]):
        
        img_patch = img[ix:ix+kernel.shape[0], ij:ij+kernel.shape[1]]
        k_prod = img_patch*kernel
        
        new_img[ix][ij] = max(0, k_prod.sum())
```

```python
plt.imshow(new_img.astype(np.uint8), cmap='gray')
plt.show()
```
