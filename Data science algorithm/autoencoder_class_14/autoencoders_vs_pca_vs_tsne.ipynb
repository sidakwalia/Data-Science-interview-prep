```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
data = pd.read_csv("/home/vasu/Downloads/fashion-mnist_test.csv").values
```

```python
X = data[:,1:]
x = data[:,1:]
x = x/255.0
x = x.reshape((x.shape[0],28,28,1))
y = data[:,0]

print (x.shape, y.shape)
```

```python
encoder = load_model("encoder.h5")
```

```python
dim_2_data = encoder.predict(x)

print (dim_2_data.shape)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(dim_2_data[:,0],dim_2_data[:,1], c=y)
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data2pca = pca.fit_transform(X)
```

```python
print (data2pca.shape)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(data2pca[:,0],data2pca[:,1], c=y)
plt.show()
```

## t-SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
```

```python
data2tsne = tsne.fit_transform(X)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(data2tsne[:,0],data2tsne[:,1], c=y)
plt.show()
```
