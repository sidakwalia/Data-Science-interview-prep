```python
f = open("./code.txt", 'r')
lr = f.read()
f.close()
```

```python
data = " ".join([ix.strip() for ix in lr.split("\n")]).lower()
```

```python
T = {}

order=4

for ix in range(len(data)-order):
    ## current context
    ctx = data[ix:ix+order]
    
    next_letter = data[ix+order]
    
    if T.get(ctx) is None:
        T[ctx] = {}
        T[ctx][next_letter] = 1
    else:
        if T[ctx].get(next_letter) is None:
            T[ctx][next_letter] = 1
        else:
            T[ctx][next_letter] += 1
```

```python
T
```

```python
for kx in T.keys():
    s = float(sum(T[kx].values()))
    
    for k in T[kx].keys():
        T[kx][k] = T[kx][k]/s
```

```python
def generate_text(ctx):
    
    possible = T.get(ctx)
    
    if possible is None:
        return " "
    
    keys = possible.keys()
    random.shuffle(keys)
    
    keys_probs = [possible[kx] for kx in keys]
    
    return np.random.choice(keys, p=keys_probs )
```

```python
ctx = "impo"

sentence = "" + ctx

for ix in range(1000):
    
    nxt = generate_text(ctx)
    sentence += nxt
    ctx = sentence[-order:]
    
print(sentence)
```
